```python
import pandas as pd

customers = pd.DataFrame({
"Customer_ID":["C001","C002","C003","C005","C006"],
"Customer_Name":["Abel","Sara","John","Helen","Samuel"],
"sex":["M","F","M","F","M"],
"City":["Addis Ababa","Adama","Jimma","Bahir Dar","Hawassa"]
})

orders = pd.DataFrame({
"Order_ID":[101,102,103,104,105,106],
"Customer_ID":["C001","C002","C003","C001","C004","C006"],
"Product":["Laptop","Phone","Keyboard","Mouse","Monitor","Tablet"],
"Amount":[1200,800,150,45,400,700]
})

employees = pd.DataFrame({
"Employee_ID":[1,2,3,4,5,6],
"Department":["IT","IT","Sales","Sales","HR","HR"],
"Employee":["Abel","Sara","John","Helen","Samuel","David"],
"Salary":[5000,6500,4500,4700,5200,4900]
})

inner_join = pd.merge(orders, customers, on="Customer_ID", how="inner")

print(inner_join)
print("Number of rows:", len(inner_join))

missing_customers = set(orders["Customer_ID"]) - set(customers["Customer_ID"])

print("Missing Customer_ID(s):", missing_customers)
```

```python
left_join = pd.merge(orders, customers, on="Customer_ID", how="left")

print(left_join)

missing_info = left_join[left_join["Customer_Name"].isna()]

print(missing_info)

filling_missing_info = left_join.fillna({"Customer_Name": "Unknown", "City": "Unknown"})
print(filling_missing_info)
```

```python
right_join = pd.merge(orders, customers, on="Customer_ID", how="right")

print(right_join)

no_orders = right_join[right_join["Order_ID"].isna()]

print(no_orders)
```

```python
outer_join = pd.merge(orders, customers, on="Customer_ID", how="outer")

print(outer_join)

print("Number of rows:", len(outer_join))

customer_only_found_in_one_table = outer_join[outer_join["Customer_Name"].isna() | outer_join["Order_ID"].isna()]
print(customer_only_found_in_one_table)
```

```python
#==========================
#filtering data
#==========================
high_amount = orders[orders["Amount"] > 500]

print(high_amount)

orders_laptop_tablet = orders[orders["Product"].isin(["Laptop", "Tablet"])]
print(orders_laptop_tablet)


orders_between = orders[orders["Amount"].between(200, 900)]

print(orders_between)
```

```python
order_from_adis_and_hawassa = customers[customers["City"].isin(["Addis Ababa", "Hawassa"])]
print(order_from_adis_and_hawassa)
```

```python
customers_name_and_city = customers[["Customer_Name", "City"]]
print(customers_name_and_city)
```

```python
name_starts_with_s = customers[customers["Customer_Name"].str.startswith("S")]
print(name_starts_with_s)
```

```python
all_products_except_Mouse = orders[orders["Product"] != "Mouse"]["Product"].unique()
print(all_products_except_Mouse)
```

```python
#===========================
#grouping and aggregation
#===========================

total_sales_amount = orders["Amount"].sum()
print(total_sales_amount)
```

```python
avarege_sales_amount = orders["Amount"].mean()
print(avarege_sales_amount)
```

```python
highest_sales_amount = orders["Amount"].max()
print(highest_sales_amount)
```

```python
lowest_sales_amount = orders["Amount"].min()
print(lowest_sales_amount)
```

```python
tolal_num_of_orders = orders["Order_ID"].count()
print(tolal_num_of_orders)
```

```python
department_salary = employees.groupby("Department")["Salary"].sum()

print(department_salary)
```

```python
department_average_salary = employees.groupby("Department")["Salary"].mean()

print(department_salary)
```

```python
max_salary_per_department = employees.groupby("Department")["Salary"].max()
print(max_salary_per_department)
```

```python
min_salary_per_department = employees.groupby("Department")["Salary"].min()
print(min_salary_per_department)
```

```python
count_employees_per_department = employees.groupby("Department")["Employee_ID"].count()
print(count_employees_per_department)
```

```python
avarage_male_customers_order_amount = orders.merge(customers[customers["sex"] == "M"], on="Customer_ID")["Amount"].mean()
print(avarage_male_customers_order_amount)
```

```python
count_of_employees_per_department = employees.groupby("Department")["Employee_ID"].count()
print(count_of_employees_per_department)
```

```python
salary_stats = employees.groupby("Department")["Salary"].agg(
    Mean_Salary="mean",
    Minimum_Salary="min",
    Maximum_Salary="max",
    Employee_Count="count"
)

print(salary_stats)
```

```python
salary_stats["Bonus"] = salary_stats["Mean_Salary"] * 0.1
print(salary_stats)
```

```python
total_salary_bonus = salary_stats.copy()
total_salary_bonus["total_salary"] = total_salary_bonus["Mean_Salary"] + total_salary_bonus["Bonus"]
print(total_salary_bonus)
```
