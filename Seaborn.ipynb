```python
# Import seaborn and apply the default theme
import seaborn as sns
sns.set_theme()
tips = sns.load_dataset("tips") # Load an example dataset
# Create a visualization
sns.relplot(data=tips, x="sex", y="total_bill", col="sex", hue="smoker", style="smoker", size="size",)
```

```python
fmri = sns.load_dataset("fmri")
sns.relplot(
 data=fmri, kind="line", x="timepoint", y="signal", col="region",
 hue="event", style="event",
)
```

```python
# Import seaborn and apply the default theme
import seaborn as sns
sns.set_theme()
tips = sns.load_dataset("tips") # Load an example dataset
sns.lmplot(data=tips, x="total_bill", y="tip", col="time", hue="smoker")
```
