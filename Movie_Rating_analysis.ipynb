```python
#==============================
#Explore the Dataset
#==============================
import pandas as pd
movies = pd.read_csv("movies_updated.csv")

print("First 5 Rows:")
print(movies.head())
print("\nDataset Information:")
movies.info()
print("\nSummary Statistics:")
print(movies.describe())
```

```python
#===================================
#Number of Movies Released EachYear
#===================================
import pandas as pd
import matplotlib.pyplot as plt

movies_per_year = movies['year'].value_counts().sort_index()
print(movies_per_year)

max_year = movies_per_year.idxmax()
max_movies = movies_per_year.max()

print(f"\nYear with the most movies: {max_year}")
print(f"Number of movies: {max_movies}")

plt.figure(figsize=(10, 5))
plt.plot(movies_per_year.index, movies_per_year.values, marker='o')
plt.title('Number of Movies Released per Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.grid(True)
plt.show()
```

```python
#==============================
#Top 10 Movie Genres
#==============================
top_genres = movies['genre'].value_counts().head(10)

print(top_genres)

plt.figure(figsize=(10, 6))
plt.bar(top_genres.index, top_genres.values)
plt.title('Top 10 Genres by Number of Movies')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.show()

print("Genre with most movies:", top_genres.idxmax())
print("Genre with fewest movies among top 10:", top_genres.idxmin())
```

```python
#==============================
#Top 10 Directors
#==============================
top_directors = movies['director'].value_counts().head(10)

print(top_directors)

plt.figure(figsize=(10, 6))
plt.barh(top_directors.index, top_directors.values)
plt.title('Top 10 Directors by Number of Movies')
plt.xlabel('Number of Movies')
plt.ylabel('Director')
plt.gca().invert_yaxis()  
plt.show()


print("Director with most movies:", top_directors.idxmax())
print("Number of movies:", top_directors.max())
```

```python
#==============================
#Distribution of Movie Ratings
#==============================
plt.figure(figsize=(10, 6))
plt.hist(movies['score'], bins=10, edgecolor='black')
plt.title('Distribution of Movie Scores')
plt.xlabel('Score')
plt.ylabel('Number of Movies')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
#==============================
#Average Movie Score by Genre
#==============================
genre_scores = movies.groupby('genre')['score'].mean()

genre_scores = genre_scores.sort_values(ascending=False)


print(genre_scores)
plt.figure(figsize=(10, 6))
genre_scores.plot(kind='bar')
plt.title('Average Movie Score by Genre')
plt.xlabel('Genre')
plt.ylabel('Average Score')
plt.xticks(rotation=45)
plt.show()


print("Highest average score genre:", genre_scores.idxmax())
print("Average score:", round(genre_scores.max(), 2))

print("Lowest average score genre:", genre_scores.idxmin())
print("Average score:", round(genre_scores.min(), 2))
```

```python
#==============================
#Top 10 Highest Grossing Movies
#==============================
import matplotlib.pyplot as plt
top10_movies = movies.nlargest(10, 'gross')

# Display movie names and revenues
print(top10_movies[['name', 'gross']])

# Create horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(top10_movies['name'][::-1], top10_movies['gross'][::-1])

plt.title('Top 10 Highest Grossing Movies')
plt.xlabel('Gross Revenue ($)')
plt.ylabel('Movie')

plt.tight_layout()
plt.show()
```
