```python
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [120, 150, 170, 160, 190]

plt.plot(
    months,
    sales,
    color='green',     
    linestyle='--',     
    marker='s',         
    markersize=8,      
    label='Sales'       
)

plt.title('Monthly Sales')
plt.legend()

plt.show()
```

```python
subjects = ['Math', 'Physics', 'Chemistry', 'Biology']
scores = [85, 78, 92, 88]

plt.plot(
    subjects,
    scores,
    color='red',
    linestyle='--',
    linewidth=3,
    marker='*',
    markersize=8,
    markerfacecolor='yellow',
    label='Scores'
)

plt.title('Exam Scores')

plt.legend()
plt.show()
```

```python
years = [2021, 2022, 2023, 2024]
revenue = [500, 700, 850, 1000]

plt.plot(
    years,
    revenue,
    color='blue',
    linestyle='--',
    linewidth=3,
    marker='d',
    markersize=8,
    markerfacecolor='lightblue',
    label='Revenue'
)

plt.title('Annual Revenue')
plt.xlabel('Year')
plt.ylabel('Revenue ($)')
plt.legend()
plt.show()
```

```python
fruits = ['Apple', 'Orange', 'Banana', 'Mango']
counts = [50, 70, 40, 60]
colors = ['red', 'orange', 'yellow', 'green']

fig, ax = plt.subplots()

ax.bar(fruits, counts, color=colors, label='Fruit Supply')

ax.set_title('Fruit Supply')
ax.set_ylabel('Quantity')

ax.legend()

plt.show()
```

```python
import numpy as np

departments = ('CS', 'IS', 'SE', 'IT')

students = {
'Male': np.array([70, 65, 80, 50]),
'Female': np.array([60, 55, 70, 45])
}

fig, ax = plt.subplots()


bars1 = ax.bar(departments, students['Male'], width=0.7, label='Male')


bars2 = ax.bar(
    departments,
    students['Female'],
    width=0.7,
    bottom=students['Male'],
    label='Female'
)


ax.bar_label(bars1, label_type='center')
ax.bar_label(bars2, label_type='center')



ax.set_title('Students by Department and Gender')



ax.legend()

plt.show()
```

```python
import pandas as pd
sales = pd.DataFrame({
    "day": ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"],
    "cups_sold": [120, 135, 110, 145, 160, 210, 190]
})

plt.plot(sales["day"], sales["cups_sold"], color='blue', marker='o', linestyle='-', markersize=8, label='Cups Sold')
plt.title('Daily Coffee Sales')
plt.xlabel('Day')
plt.ylabel('Cups Sold')
plt.legend()
plt.show()
```
