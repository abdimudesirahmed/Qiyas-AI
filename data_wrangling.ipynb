```python
#=================================
#Bank Loan Approval Dashboard
#=================================
data = {
'Region':['Addis','Addis','Adama','Adama','Hawassa','Bahir Dar'],
'Loan_Type':['Personal','Mortgage',
'Personal','Mortgage',
'Personal','Mortgage'],
'Status':['Approved','Rejected',
'Approved','Approved',
'Rejected','Approved']

}

df = pd.DataFrame(data)
df['Count'] = 1


pivot = pd.pivot_table(
    df,
    values='Count',
    index='Region',
    columns=['Loan_Type', 'Status'],
    aggfunc='sum'
)


pivot = pivot.fillna(0)

print("Pivot Table:")
print(pivot)

approved_loans = (df['Status'] == 'Approved').sum()
print("\nTotal Approved Loans:", approved_loans)


approved_by_region = df[df['Status'] == 'Approved'].groupby('Region').size()

top_region = approved_by_region.idxmax()
max_approved = approved_by_region.max()

print("\nRegion with highest approved loans:", top_region)
print("Number of approved loans:", max_approved)
```

```python
#=================================
#Telecom Customer Usage Dashboard
#=================================

data = {
'Customer_ID':['C1','C1','C2','C2','C1','C3'],
'Month':['Jan','Jan','Jan','Jan','Feb','Feb'],
'Service':['Data','Voice','Data','Voice','Data','Voice'],
'Usage_GB':[12,5,8,10,np.nan,15]
}

df = pd.DataFrame(data)
print("Missing Values:")
print(df.isnull().sum())

avg_usage = df['Usage_GB'].mean()
df['Usage_GB'] = df['Usage_GB'].fillna(avg_usage)


pivot = pd.pivot_table(
    df,
    values='Usage_GB',
    index='Customer_ID',
    columns=['Month', 'Service'],
    aggfunc='sum'
)


pivot = pivot.fillna(0)

print("\nPivot Table:")
print(pivot)


total_usage = df.groupby('Customer_ID')['Usage_GB'].sum()

highest_customer = total_usage.idxmax()
highest_usage = total_usage.max()

print("\nCustomer with Highest Total Usage:")
print(highest_customer, "-", highest_usage)


avg_customer_usage = df.groupby('Customer_ID')['Usage_GB'].mean()

print("\nAverage Usage per Customer:")
print(avg_customer_usage)
```

```python
#===================================
#Product Code Cleaning and Analysis
#===================================

data = {
'Product_Code':[
'TV-2025-001',
'LAP-2025-017',
'PHN-2024-023',
'TAB-2025-011'
],
'Price':['45000','65000','32000','55000']
}

df = pd.DataFrame(data)
df['Price'] = df['Price'].astype(int)

df[['Category', 'Year', 'Product_Number']] = df['Product_Code'].str.split('-', expand=True)
average_price = df['Price'].mean()
most_expensive = df.loc[df['Price'].idxmax()]

products_2025 = df[df['Year'] == '2025']


total_sales = df['Price'].sum()

print("Updated DataFrame:")
print(df)

print("\nAverage Product Price:", average_price)

print("\nMost Expensive Product:")
print(most_expensive)

print("\nProducts Released in 2025:")
print(products_2025)

print("\nTotal Sales Value:", total_sales)
```

```python
#=======================================
#Human Resource Attendance Analysis
#=======================================
data = {
'Employee_ID':['E1','E1','E1','E2','E2','E3','E3'],
'Month':['Jan','Jan','Feb','Jan','Feb','Jan','Feb'],
'Status':['Present',
'Absent',
'Present',
'Present',
'Present',
'Absent',
'Absent']

}

df = pd.DataFrame(data)
df['Count'] = 1
pivot = pd.pivot_table(
    df,
    values='Count',
    index='Employee_ID',
    columns=['Month', 'Status'],
    aggfunc='sum'
)


pivot = pivot.fillna(0)

print("Pivot Table:")
print(pivot)


absences = df[df['Status'] == 'Absent'].groupby('Employee_ID').size()
highest_absent_employee = absences.idxmax()
total_records = df.groupby('Employee_ID').size()


absence_percentage = (
    df[df['Status'] == 'Absent']
    .groupby('Employee_ID')
    .size()
    .reindex(total_records.index, fill_value=0)
    / total_records * 100
)

attendance_rate = 100 - absence_percentage
best_employee = attendance_rate.idxmax()

print("\nTotal Absences:")
print(absences)

print("\nEmployee with Highest Absences:")
print(highest_absent_employee)

print("\nTotal Attendance Records:")
print(total_records)

print("\nAbsence Percentage:")
print(absence_percentage)

print("\nBest Attendance Rate:")
print(best_employee)
```

```python
#==============================
#Customer Orders Analysis
#==============================
data = {
'Customer_ID':['C001','C002','C003','C003'],
'Full_Name':['Abebe Kebede',
' Hana Alemu ',
'Meron Tadesse',
'Meron Tadesse'],

'Address':[
'Bole Road, Addis Ababa, Bole, 03, H125',
'CMC Road, Addis Ababa, Yeka, 10, H201',
'Piassa, Addis Ababa, Arada, 01, H505',
'Piassa, Addis Ababa, Arada, 01, H505'
],
'Product':['Laptop','Phone','Laptop','Tablet'],
'Amount':[45000,np.nan,32000,28000]
}

df = pd.DataFrame(data)
print("Missing Values:")
print(df.isnull().sum())

avg_amount = df['Amount'].mean()
df['Amount'] = df['Amount'].fillna(avg_amount)
df = df.drop_duplicates()
df['Full_Name'] = df['Full_Name'].str.strip()
df[['First_Name', 'Last_Name']] = df['Full_Name'].str.split(' ', expand=True)
df[['Street', 'City', 'Subcity', 'Woreda', 'House_No']] = (
    df['Address'].str.split(',', expand=True)
)
for col in ['Street', 'City', 'Subcity', 'Woreda', 'House_No']:
    df[col] = df[col].str.strip()
df['Count'] = 1
pivot = pd.pivot_table(
    df,
    values='Count',
    index='Customer_ID',
    columns='Product',
    aggfunc='sum',
    fill_value=0
)

print("\nPivot Table:")
print(pivot)
customer_spending = df.groupby('Customer_ID')['Amount'].sum()

top_customer = customer_spending.idxmax()
max_spent = customer_spending.max()

print("\nCustomer Who Spent Most:")
print(top_customer, "-", max_spent)
average_purchase = df['Amount'].mean()

print("\nAverage Purchase Amount:")
print(average_purchase)
above_average = df[df['Amount'] > average_purchase]

print("\nCustomers Above Average Purchase:")
print(above_average)
df['Full_Address'] = (
    df['Street'] + ', ' +
    df['City'] + ', ' +
    df['Subcity'] + ', ' +
    df['Woreda'] + ', ' +
    df['House_No']
)

df = df.drop(
    columns=['Street', 'City', 'Subcity', 'Woreda', 'House_No']
)
print("\nFinal Cleaned Dataset:")
print(df)
```
