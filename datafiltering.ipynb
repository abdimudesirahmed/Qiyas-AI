```python
import pandas as pd
data = {
 'Order_ID': ['1001', '1002', '1003', '1004', '1005'],
 'Product_ID': ['P101', 'P102', 'P103', 'P101', 'P104'],
 'Product_Name': ['Laptop', 'Smartphone', 'Headphones', 'Laptop', 
'Coffee Maker'],
 'Category': ['Electronics', 'Electronics', 'Accessories', 
'Electronics', 'Kitchen'],
 'Quantity': [2, 1, 5, 3, 2],
 'Amount': [1600, 600, 500, 2400, 100],
 'Date': ['2024-01-10', '2024-01-11', '2024-01-12', '2024-01-12', 
'2024-01-13'],
 'Region': ['North', 'South', 'East', 'West', 'North'],
}
df_sales = pd.DataFrame(data)
df_sales
```

```python
ds = df_sales[((df_sales['Region'] == 'East') | (df_sales['Region'] == 'South')) & (df_sales['Amount'] >= 500)]
print(ds)
```

```python
sum_per_product = df_sales.groupby('Category')['Amount'].sum()
print(sum_per_product)
```

```python
df_sales["Total_Amount"] = df_sales["Quantity"] * df_sales["Amount"]
df_sales["Tax"] = df_sales["Total_Amount"] * 0.15
df_sales["Net_Amount"] = df_sales["Total_Amount"] - df_sales["Tax"]
print(df_sales)
```

```python
df_sales["Date"] = pd.to_datetime(df_sales["Date"])
jan_10_11 = df_sales[
    (df_sales["Date"] >= "2024-01-10") &
    (df_sales["Date"] <= "2024-01-11")
]

print(jan_10_11)
```

```python
import pandas as pd
data = {
 'Student_ID': ['S001', 'S002', 'S003', 'S004', 'S005'],
 'Mathematics' : [85, 75, 95, 65, 90],
 'Science' : [78, 88, 82, 85, 79],
 'English' : [90, 85, 80, 70, 95],
 'Information Technology' : [89, 65, 88, 92, 98]
}
df_result = pd.DataFrame(data)
df_result
```

```python
import numpy as np
df_result['Average'] = np.mean(df_result[['Mathematics', 'Science', 
'English', 'Information Technology']], axis=1)
df_result
```

```python
df_result['Weighted Average'] = (df_result['Mathematics'] * 0.4) + \
(df_result['Science'] * 0.2) + (df_result['English'] * 0.2) + \
(df_result['Information Technology'] * 0.2)
df_result
```

```python
def standardize(column):
 mean = np.mean(column)
 std_dev = np.std(column)
 return (column - mean) / std_dev
# Standardize the columns
df_result['Math_Std'] = standardize(df_result['Mathematics'])
df_result['Sci_Std'] = standardize(df_result['Science'])
df_result['Eng_Std'] = standardize(df_result['English'])
df_result['IT_Std'] = standardize(df_result['Information Technology'])
df_result
```

```python
def assign_grade(weighted_avg):
	if weighted_avg >= 85:
		return 'A'
	elif weighted_avg >= 75:
		return 'B'
	elif weighted_avg >= 65:
		return 'C'
	elif weighted_avg >= 55:
		return 'D'
	else:
		return 'F'

# Apply the grading function to the weighted averages
df_result['Final_Grade'] = df_result['Weighted Average'].apply(assign_grade)
df_result
```
